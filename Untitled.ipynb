```python
EN_PR = False # False -> stable
RANDOM_ROUND = 1000
DIGITS = 6
SAMPLE_RATIO = 0.03
PLOT_THINGS = True
GRAPH_EVERY = 100
BAD_CHOICE_RATIO = 0.15
```

```python
import math
from collections import Counter
import random
from tqdm import tqdm_notebook as tqdm
from scipy.stats import kurtosis
```

```python
# !pip install --user seaborn
%matplotlib inline
def savefig(fig_idx={}, fig_name='yes-animation-frame'):
    if fig_name not in fig_idx:
        fig_idx[fig_name] = [0]
    fig_idx = fig_idx[fig_name]
    plt.savefig(f'{fig_name}-{fig_idx[0]:04}.png')
    fig_idx[0] += 1

def kurtosisf(ds:dict, frame=[0]):
    plt.figure()
    for d in ds:
        sns.regplot(list(ds[d].keys()), list(ds[d].values()))
    plt.title(frame[0])
    frame[0] += 1
    savefig(fig_name='kurtosis-animation-frame')
    return plt
```

```python
if PLOT_THINGS:
    # bring out the disgusting guns
    import numpy as np
    import seaborn as sns
    import matplotlib.pyplot as plt
```

```python
def entropy(s):
    p, lns = Counter(s), float(len(s))
    return -sum( count/lns * math.log(count/lns, 2) for count in p.values() if count != 0)
```

```python
tspace=10**DIGITS
samples=int(tspace*SAMPLE_RATIO)
numbers = random.choices(range(tspace), k=samples)
```

```python
numbers[373] = 655556  # set it to something out of that distribution
secret = numbers[373]
def secret_test(x):
    return x == secret
print(secret) # we need to know!
```

```python
passlen = DIGITS
```

```python
if PLOT_THINGS:
    # draw initial distributions
    plt.figure()
    for digitx in range(passlen):
        sns.distplot([int(str(x).zfill(DIGITS)[digitx]) for x in numbers], hist=False, rug=True, label=f'10^{DIGITS-1-digitx}')
    plt.title('initial distribution')
    savefig()
    # pretty uniform
```

```python
def gnumber(srt, x):
    try:
        # print(srt, x)
        return tuple([next(y[1] for y in srt[i] if y[0]==x) for i,x in enumerate(int(x) for x in str(x))])
    except:
        return (99999,)*passlen
```

```python
def decide_one_round(numbers:list, digits:dict):
    # sort by entropy in digits
    numbers = numbers[:]
    digits = digits.copy()
    items = list(digits.items())
    random.shuffle(items)
    srt = sorted(items, key=lambda x: kurtosis(np.array(list(x[1].values()))), reverse=not EN_PR)
    # srt = dict((x[0], sorted(x[1].items(), key=lambda y: y[1], reverse=EN_PR if random.choice(range(100))/100>BAD_CHOICE_RATIO else not EN_PR)) for x in srt)
    number = (min if EN_PR or random.choice(range(100))/100<BAD_CHOICE_RATIO else max)(
        numbers,
        key = lambda x: gnumber(srt, x)
    )
    for i,x in enumerate(int(x) for x in str(number)):
        digits[i][x] -= 1
    digits = {x:{u:v for u,v in y.items() if v != 0} for x,y in digits.items()}
    numbers.remove(number)
    return number, numbers, digits
```

```python
n = None
try:
    ns = numbers
    for ii in tqdm(range(RANDOM_ROUND)):
        n = random.choice(ns)
        if secret_test(n):
            print('lucky success in init round', ii+1)
            # break
        else:
            ns.remove(n)
    if PLOT_THINGS:
        # vis after a bunch of crap is removed
        # this knocks the dists a bit farther from each other
        plt.figure()
        for digitx in range(passlen):
            sns.distplot([int(str(x).zfill(DIGITS)[digitx]) for x in ns], hist=False, rug=True, label=f'10^{DIGITS-1-digitx}')
        plt.title('post init rounds')
        savefig()
except Exception as e:
    print(e)
```

```python
digits = {x:{y:0 for y in range(10)} for x in range(passlen)}
for number in numbers:
    s = str(number)
    for x in range(min(passlen, len(s))):
        digits[x][int(s[x])] += 1
ds = {x:{u:v for u,v in y.items() if v != 0} for x,y in digits.items()}
```

```python
#try:
count = 0
with tqdm(total=len(ns)) as tq:
    while not secret_test(n):
        tq.set_description(desc=str(n).zfill(DIGITS)+'('+str(secret).zfill(DIGITS)+')', refresh=False)
        tq.update()
        n,ns,ds = decide_one_round(ns, ds)
        count = tq.n
        if PLOT_THINGS and tq.n%GRAPH_EVERY==0:
            plt.clf()
            for digitx in range(passlen):
                sns.distplot([int(str(x).zfill(DIGITS)[digitx]) for x in ns], hist=False, rug=True, label=f'10^{DIGITS-1-digitx}')
            plt.title('iteration ' + str(tq.n))
            savefig()
            plt.show()
            kurtosisf(ds).show()
print('success! found', n, 'in', count, 'iteration(s) of decide_one_round')
#except Exception as e:
#    print(e)
```
